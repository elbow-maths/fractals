```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

from koch_snowflake import *
```

This notebook uses the koch_snowflake.py code to generate lovely snowflakes.

The code has a very simple function called koch_snowflake and you just enter the number of iterations of the process you would like it to undertake before returning the image.

The snowflake starts as a equilateral triangle, so if we plot a snowflake with 0 iterations we get:

```python
koch_snowflake(0)
```

We can then see the result if we perform an iteration of the process to the triangle:

```python
koch_snowflake(1)
```

The middle third of each line of the triangle has been changed to be an equilateral triangle (without a base)

If we increase the iterations again we can see what it looks like after two iterations:

```python
koch_snowflake(2)
```

```python
koch_snowflake(3)
```

```python
koch_snowflake(4)
```

You could continue to increase the number of iterations and the edges of the snowflake would become increasingly complex. However, the number of calculations needed at each iteration grows exponentially so it's a good idea not to do too many!
